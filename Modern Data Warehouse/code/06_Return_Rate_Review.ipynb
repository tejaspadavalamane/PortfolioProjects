```python
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
%%sql

create table if not exists TMDW_FinalProj.Gold_OrderReturn
(
    
OID  string,
Returned string,
Order_Year int,
Order_Month int,
Created_TS TIMESTAMP,
Modified_TS TIMESTAMP

)
using delta 
partitioned by(Order_Year,Order_Month)
```

```python
Maximum_Date=spark.sql("select coalesce(max(Modified_TS),'1900-01-01') from TMDW_FinalProj.Gold_OrderReturn").first()[0]
```

```python
#checking what we will get in Maximum_Date Variable
Maximum_Date
```

```python
dataframe =spark.sql("""select OID  ,
Return ,
Order_Year, 
Order_Month ,
Created_TS ,
Modified_TS 
from TMDW_FinalProj.BronzeSales where Return="Yes" and Modified_TS>'{}'""".format(Maximum_Date))
```

```python
# creating a View for the data
dataframe.createOrReplaceTempView("View_Return")
```

```python
%%sql
insert into TMDW_FinalProj.Gold_OrderReturn
select * from View_Return
```
