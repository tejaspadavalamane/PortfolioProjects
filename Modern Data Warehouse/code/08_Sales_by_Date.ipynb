```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
%%sql
create table if not EXISTS TMDW_FinalProj.gld_date
(
BDate Date,
BYear int,
BMonth int,
BQuarter int,
BWeek int
)
using DELTA
PARTITIONED by (BYear)
```

```python
start_date='2015-01-01'
end_date='2030-12-31'

date_difference=spark.sql("select date_diff('{}','{}')".format(end_date,start_date)).first()[0]
```

```python
date_difference
```

```python
ID_Data=spark.range(0,date_difference)
```

```python
date_data=ID_Data.selectExpr("date_add('{}',cast(id as int)) as BDate".format(start_date))
```

```python
date_data.createOrReplaceTempView("View_Date")
```

```python
%%sql

delete from TMDW_FinalProj.gld_date;
```

```python
%%sql
insert into TMDW_FinalProj.gld_date
select *, 
Year(BDate) as BYear,
Month(BDate) as BMonth,
Quarter(BDate) as BQuarter,
weekday(BDate) as BWeek
 from View_Date
```
